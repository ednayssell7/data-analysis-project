```python
#import data
import pandas as pd
import numpy as np
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import string
import collections

import matplotlib.pyplot as plt

filename = 'datasets/articles_data_homeproject.csv'

#read a CSV file with pandas
df = pd.read_csv(filename)
```

```python
words = df['title'].str.split(' ', n=50, expand=True)
words.drop_duplicates(keep=False,inplace=True) 
words = df.to_string(index = False) 

print(type(words))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["news", "reuters", "Reuter", "https", "image", "said", "reuters", "id", "html", "super", "time", "bbc", "cnn", "new", "says", "to",  "is", "still", "with", "of", "&", "for",  "be", "from", "have", "had", "and", "None" ])

wordcloud = WordCloud(stopwords=stopwords, max_words=400).generate(words)

plt.figure(figsize = (15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("img/first_word_cloud.png")
```
